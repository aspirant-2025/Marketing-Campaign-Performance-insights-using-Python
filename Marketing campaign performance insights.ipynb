```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#scatterplot
plt.figure(figsize=(10,5))
sns.scatterplot(x='Acquisition_Cost', y='ROI', data=df)
plt.xlabel('Acquisition Cost')
plt.ylabel('ROI')
plt.show()
```

```python
group = df.groupby(['Channel_Used', 'Campaign_Type'])['Conversion_Rate'].mean().reset_index()
```

```python
#barchart
plt.figure(figsize=(10,5))
sns.barplot(data=group, x='Channel_Used', y='Conversion_Rate', hue='Campaign_Type')
plt.ylabel('Average Conversion Rate')
plt.xlabel('Channel Used')
plt.legend(title='Campaign_Type', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

```python
#boxplot
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='Campaign_Type', y='Engagement_Score')
plt.xlabel('Campaign Type')
plt.ylabel('Engagement Score')
plt.show()
```

```python
avg_ROI = df.groupby('Company')['ROI'].mean().reset_index()
avg_ROI
```

```python
# Average ROI by Company using barchart
plt.figure(figsize=(10,6))
sns.barplot(data=avg_ROI, x='Company', y='ROI', color='lightblue')
plt.title('Average ROI by Company')
plt.xlabel('Company')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.show()


```

```python
correlation = df[['Engagement_Score', 'Conversion_Rate']].corr()
correlation

```

```python
sns.heatmap(correlation)
```

```python
#boxplot
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Customer_Segment', y='Acquisition_Cost', hue='Channel_Used')
plt.xlabel('Customer Segment')
plt.ylabel('Acquisition Cost')
plt.legend(title='Channel Used', bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

```python
avg = df.groupby('Language')['Conversion_Rate'].mean().reset_index()
avg
```

```python
#boxplot
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x='Customer_Segment',y='Acquisition_Cost',hue='Channel_Used')
plt.xlabel('Customer Segment')
plt.ylabel('Acquisition Cost')
plt.xticks(rotation=45)
plt.legend(title='Channel Used', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

**Channel Effectiveness**

```python
#barplot
plt.figure(figsize=(10,5))
sns.barplot(data=df, x='Channel_Used', y='Engagement_Score', hue='Campaign_Type')
plt.xlabel('Channel Used')
plt.ylabel('Engagement Score')
plt.legend(title='Campaign Type' , bbox_to_anchor=(1,1) , loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#piechart
ROI_by_channel = df.groupby('Channel_Used')['ROI'].sum().reset_index()
plt.figure(figsize=(5,5))
plt.pie(ROI_by_channel['ROI'], labels=ROI_by_channel['Channel_Used'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')  
plt.show()
```

```python
#scatterplot
sns.scatterplot(data=df, x='Impressions', y='Clicks', hue='Channel_Used', palette='Set2')
```
